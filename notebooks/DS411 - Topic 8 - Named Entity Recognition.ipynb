# __Named Entity Recognition__

<font size='4px'><b>Named Entity Recognition</b></font> (NER) is a fundamental process in identifying and classifying key information (*entities*) in text into predefined categories like names, places, organizations, dates, and so on.<br>

`Named Entities` are words or phrases that refers to real-world objects, such as:
- `PER` - Person (e.g Isaac Newton, Albert Einstein, etc.)
- `ORG` - Organization (e.g Google, Apple, NASA, etc.)
- `GPE/LOC` - Geopolitical Entities/Location (e.g Cagayan de Oro City, USTP, etc.)
- `DATE` - Dates

Amongst others.

An example of this is:

> I am Sean and I go at USTP.

NER would then identify and tag it as follows:<br>
> I am [PER]Sean and I go at [LOC]USTP.

<font size='4px'><b>Named Entity Recognition</b></font> is a **preprocessing** and **feature extraction** step that feeds useful information into predictive models. It is widely used in information retrieval, question answering systems, and entity linking tasks.

# NER with NLTK

<font size='4px'><b>NLTK</b></font> comes with built-in NER module that utilizes a classifier trained on a labeled dataset to identify named entities in a textual data.

```python
import nltk
from nltk import pos_tag, ne_chunk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
```

## **Tokenize** your text first.

```python
text = "Apple is planning to open a new office in Manila by March 2025. Tim Cook said the investment will be around $10 Million."
tokens = word_tokenize(text)
print(tokens)
```

## **POS Tagging**
After ___Tokenizing___ our text, next is to do __POS Tagging__.

```python
tags = pos_tag(tokens)
tags
```

> *You need POS tagging before NER because it relies on how words function within a sentence (their parts of speech) to correctly identify entities.*

## NER

```python
ner_words = ne_chunk(tags)
print(ner_words)
```

# So how does POS Tagging help NER

POS tagging acts as a foundation for NER — it provides grammatical structure, helps locate and interpret proper nouns, and offers context for accurate entity classification.

1. **Identifying Proper Nouns**
   - POS tagging labels words based on their grammatical role, such as noun, verb, adjective, etc.
In NER, we often focus on proper nouns (tagged as NNP or NNPS in English), because they usually represent names of people, organizations, or places.
> “Barack Obama was born in Hawaii.”
POS tagging marks “Barack Obama” and “Hawaii” as proper nouns — strong candidates for named entities.
   
2. **Disambiguating Words**
    - Some words can have different meanings depending on their role in a sentence. POS tagging helps clarify that role.
This helps NER systems understand whether a word is being used as a name or a common term.
> “Apple released a new product.” (Proper noun — company)<br>
“She ate an apple.” (Common noun — fruit)<br>
POS tagging differentiates these cases, preventing wrong entity detection.

3. **Identifying Entity Boundaries**
   - NER often needs to detect where a named entity starts and ends in a sentence. POS tags help mark continuous proper nouns that form a complete name.
> “University of the Philippines”<br>
All three words are tagged as part of a proper noun phrase, signaling that they belong to one entity rather than separate ones.

5. Contextual Clues for Entity Type
   - POS tags also provide context that helps determine what kind of entity something might be. For instance, if a proper noun follows words like “Mr.”, “Dr.”, or “President”, it’s likely a person’s name.
> “Dr. Jose Rizal wrote many novels.” <br>
Here, POS tagging helps NER recognize “Jose Rizal” as a person entity because of the title “Dr.” and the proper noun tags.
