## Read PDF Files

````python
with pdfplumber.open("./Datasets/samplePDF_data.pdf") as pdf:
    tb = []
    for i in tqdm(range(len(pdf.pages))):
        page = ' '.join(pdf.pages[i].extract_text().split('\n')[1:-1])
        tb.append(page)
````

- `len(pdf.pages)` : This gets the total number of pages in the PDF.
- `tqdm(range(...))` : tqdm is a library that provides a progress bar for loops. Here, it shows the progress of the loop as it processes each page.
- `for i in ` : This loop iterates over each page in the PDF, using i as the index for each page.

The code first extracts the text from the current page of the PDF using `pdf.pages[i].extract_text()`. This extracted text is typically a single string that represents the entire content of that page. To break it down further, the `split('\n')` method is then applied to this string, splitting the text into a list of individual lines based on the <u>newline characters (\n)</u>. Each element in this list corresponds to a line of text from the page. Next, the slicing operation `[1:-1]` is used to <u>remove the first and last lines from this list</u>, likely to discard unnecessary headers and footers. Finally, the remaining lines are joined back together into a single string using `' '.join(...)`, which inserts a space between each line. This processed string, which now contains the core content of the page without the first and last lines, is then ready to be added to the tb list.

## Read HTML
----

<font size='5px'><b>What is Beautiful Soup?</b></font>
<p>
    Beautiful Soup is a Python library used for parsing HTML and XML documents. Imagine you have a webpage (HTML), and you want to pull out specific pieces of information from it, like titles, links, or images. Beautiful Soup helps you "soup" through the messy HTML code and extract just what you need.
</p>

<font size='5px'><b>How Does Requests Fit into This?</b></font>
<p>
    While Beautiful Soup is great at parsing and navigating HTML, it doesn't actually fetch the HTML content for you. That's where Requests comes in.
    Requests is another Python library, and it's used to make HTTP requests, like when your browser requests a webpage. It allows you to send a request to a server
    (a website) and get the HTML content of that page as a response.
</p>

<font size='5px'><b>How They Work Together:</b></font>

1. `Requests` fetches the webpage.
2. `Beautiful Soup` parses the webpage and helps you extract the data you want.

````python
# create a request
"""
 This session helps to keep track of things like cookies if you were to log in somewhere.
 
"""
a = requests.Session()
````

````python
"""
    Place your designated URL
"""

url = 'https://en.wikipedia.org/wiki/2024_Summer_Olympics'
````

````python
# Using Try-Except to check is we have succesfully fetch our webpage
try: 
    """
        Fetch the Webpage
    """

    response = a.get(url)
    print(response.status_code,response.reason)

except response.status_code != '202':
    """
        Checks if the webpage was successfully fetched by looking at the status code
    """
    print(response.reason)
````

````python
"""
   Using Beautiful Soup, we parse the webpage’s HTML structure. 
   We then specify which part of the HTML we want to find, such as <p> tags for paragraphs.
"""

soup = BeautifulSoup(response.text)
soup.body.find_all('p')
# soup
````

````python
"""
    It turns the paragraphs into a long string of text and then splits 
    it up so that each paragraph is separated.
"""

text = str(soup.body.find_all('p'))
text_list = text.split('<p>')[1:-1]
text_list[1]
````

````python
"""
    Here's another method to get your text with less noise
"""

text = soup.body.find_all('p')
text_list = [sentence.get_text() for sentence in text][1:-1]
text_list
````
