**Create a list of dictionaries representing all movies across different years.**

```python
movie_list = list()

for year in tqdm(movie_year):
    year_path = path / str(year)
    for movies in year_path.iterdir():
        with open(year_path / movies, encoding="ISO-8859-1") as movie_jsn:
            movie_list.append(json.load(movie_jsn))
        
```

```python
movie_list
```

**How many movies are there?**

```python
len(movie_list)
```

**Display all movies**

```python
movie_title = list()

for title in movie_list:
    movie_title.append(title['name'])
movie_title
```

**---------------------------- or this method ----------------------------**

**Show all categories per movie**

```python
movie_categories = {}

for movie in movie_list:
    if 'name' in movie and 'categories' in movie:
        if movie['name'] in movie_categories:
            movie_categories[movie['name']].extend(movie['categories'])
        else:
            movie_categories[movie['name']] = movie['categories']
```

```python
movie_categories
```

**Show the storyline of each movie**

```python
movie_storyline = {}

for movie in movie_list:
    if 'name' in movie and 'storyline' in movie:
        if movie['name'] in movie_storyline:
            movie_storyline[movie['name']] = (movie['storyline'])
        else:
            movie_storyline[movie['name']] = movie['storyline']
```

```python
with pdfplumber.open(path.parent / "sampledata_PDF.pdf") as pdf:
    tb = []
    for i in tqdm(range(len(pdf.pages))):
        page = ' '.join(pdf.pages[i].extract_text().split('\n')[1:-1])
        tb.append(page)
```

**Create a session object using the request library**

```python
session = requests.session()
```

**Grab your HTML**

```python
url = 'https://medium.com/@CloverDX/8-dimensions-of-data-quality-8cc31648c530'
```

```python
try: 
    """
        Fetch the Webpage
    """

    response = session.get(url)
    print(response.status_code,response.reason)

except response.status_code != '202':
    """
        Checks if the webpage was successfully fetched by looking at the status code
    """
    print(response.reason)
```

**Parse the HTML Page and place it in a list**

```python
soup = BeautifulSoup(response.text)
text = soup.find_all('h1')
text_list = [sentence.get_text(strip=True) for sentence in text][1:-1]
text_list
```

```python
h1_p_dict = {}

h1_tags = soup.find_all('h2')

for h1 in h1_tags:
    h1_text = h1.get_text(strip=True)
    p_tag = h1.find_next('p')
    
    if p_tag:
        p_text = p_tag.get_text(strip=True)
        h1_p_dict[h1_text] = p_text
```
